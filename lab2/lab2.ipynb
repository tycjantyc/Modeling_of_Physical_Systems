```python
from lab2 import Object
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
#alumina object
obj = Object(resolution=30, K = 237*1000, cw = 900, p = 2700, dt = 0.0005, time = 2000)  #minimal dt = 0.25*min(dx, dy)**2*(cw*p)/K/10                   
#obj = Object(resolution=30, K = 401*1000, cw = 380, p = 8920, dt = 0.01, time = 2000)  # dt = 0.25*0.01**2*self.cw*self.p/self.K/10
#obj = Object(resolution=30, K = 58*1000, cw = 450, p = 7860, dt = 0.01, time = 2000)
plt.imshow(obj.mode)
```

```python
num = obj.heat()

time = num*obj.dt

print(f"Time of heating - {time} seconds")
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

X, Y = np.meshgrid(range(obj.resolution), range(obj.resolution))
Z = obj.shape

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

ax.zaxis.set_major_locator(LinearLocator(10))

ax.zaxis.set_major_formatter('{x:.02f}')
ax.set_title("Heat transfer of alumina")


fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
